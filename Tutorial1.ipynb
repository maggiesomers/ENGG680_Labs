```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random 2D (x, y) values
num_points = 100  # Number of points
x_values = np.random.rand(num_points)  # Generate x values between 0 and 1
y_values = np.random.rand(num_points)  # Generate y values between 0 and 1
# Print the values
print("x values:", x_values)
print("y values:", y_values)
```

```python
def plot_test_set(set_x, set_y):
  plt.scatter(set_x, set_y, color='blue', marker='o')
  plt.title('Random 2D Points')
  plt.xlabel('x [unitless]')
  plt.ylabel('y [unitless]')
  plt.grid(True)
  plt.show()
```

```python
plot_test_set(x_values, y_values)
```
